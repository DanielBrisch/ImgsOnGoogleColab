```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('noruega.jpg') #carrega a imagem
print('Original', img.shape) #exibe dimensao da imagem
cv2_imshow(img)

img_resized = cv2.resize(img, (214, 362), interpolation=cv2.INTER_AREA)
print('Redimensionada:', img_resized.shape)
cv2_imshow(img_resized)
```

```python
scale = 60
width = int((img.shape[1] * scale) / 100)
height = int((img.shape[0] * scale) / 100)
print(img.shape)
print(width, height)
img_resized = cv2.resize(img, (width, height), interpolation=cv2.INTER_AREA)
cv2_imshow(img_resized)
```

```python
image = cv2.flip(img_resized, 0) #flip vertical
cv2_imshow(image)

image = cv2.flip(img_resized, 1) #flip horizontal
cv2_imshow(image)

image = cv2.flip(img_resized, -1) #flip ambos
cv2_imshow(image)
```

```python
import numpy as np
kernel = np.ones((7, 7), np.float32)/10 #Criamos o kernel
print(kernel)
conv = cv2.filter2D(img, -1, kernel) #Convolução, filtro 2D
cv2_imshow(conv)
```

```python
cinza = cv2.imread('noruega.jpg', 0)
# cv2_imshow(cinza)

blur = cv2.blur(cinza, (5, 5))
# cv2_imshow(blur)

gaussian_blur = cv2.GaussianBlur(cinza, (5,5), 0)
cv2_imshow(gaussian_blur)

median_blur = cv2.medianBlur(cinza, 5)
# cv2_imshow(median_blur)

bilateral_filter = cv2.bilateralFilter(cinza, 9, 75, 75)
# cv2_imshow(bilateral_filter)
```

```python
edges = cv2.Canny(cinza, 100, 200)
cv2_imshow(edges)
```
